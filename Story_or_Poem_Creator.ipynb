```python
import random
import re
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
from collections import Counter
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
import spacy
```

```python
# Download required resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')

# Load spaCy model
nlp = spacy.load("en_core_web_sm")
```

```python
# STEP 1: Keyword Input & Preprocessing
keywords = ["moon", "adventure", "magic", "forest", "mystery"]

def preprocess_keywords(keywords):
    stop_words = set(stopwords.words('english'))
    return [word.lower().strip() for word in keywords if word.lower() not in stop_words]

clean_keywords = preprocess_keywords(keywords)
print("\nCleaned Keywords:\n", clean_keywords)
```

```python
# STEP 2: Generate a Rule-Based Story/Poem
story_templates = [
    "Under the {noun}, a {adj} {noun} embarked on a {adj} journey.",
    "They discovered a {noun} filled with {adj} wonders.",
    "The {noun} whispered secrets of {adj} realms.",
    "With {noun} in hand, they ventured into the {noun}.",
    "In the heart of the {noun}, magic {verb} through the air."
]

poem_templates = [
    "The {noun} shines so {adj} in the sky,",
    "Through the {noun} the wind {verb} by.",
    "A {noun} of dreams, a {adj} night,",
    "Lost in the {noun}, a {adj} sight."
]

# Word Categories
nouns = ["moon", "adventure", "magic", "forest", "mystery", "castle", "river", "wind"]
adjectives = ["mystical", "dark", "glowing", "whispering", "hidden"]
verbs = ["glowed", "whispered", "danced", "swayed", "rushed"]

def generate_text(template_list):
    generated = []
    for template in template_list:
        text = template.format(
            noun=random.choice(nouns),
            adj=random.choice(adjectives),
            verb=random.choice(verbs)
        )
        generated.append(text)
    return " ".join(generated)

generated_story = generate_text(story_templates)
generated_poem = generate_text(poem_templates)

print("\n Generated Story:\n", generated_story)
print("\n Generated Poem:\n", generated_poem)
```

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize

# Ensure the necessary tokenizer is available
nltk.download('punkt')

# Sample generated story
generated_story = "The moon glowed over the mystical forest. A whispering wind carried secrets through the night."

# Tokenization (Word & Sentence)
words = word_tokenize(generated_story)
sentences = sent_tokenize(generated_story)

# Display results
print("\n Tokenized Words:\n", words)
print("\n Tokenized Sentences:\n", sentences)
```

```python
# âœ… STEP 4: Stopword Removal & Comparison
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word.lower() not in stop_words]

print("\nðŸš« Stopword Removed Words:\n", filtered_words)
```

```python
#STEP 5: POS Tagging & Named Entity Recognition (NER)
pos_tags = nltk.pos_tag(words)
doc = nlp(generated_story)
entities = [(ent.text, ent.label_) for ent in doc.ents]

print("\n POS Tagging:\n", pos_tags)
print("\n Named Entities:\n", entities)
```

```python
#STEP 6: Word Frequency Analysis & Visualization
word_counts = Counter(filtered_words)
top_words = word_counts.most_common(10)

plt.figure(figsize=(8, 5))
sns.barplot(x=[w[0] for w in top_words], y=[w[1] for w in top_words])
plt.title("ðŸ”¢ Top 10 Word Frequency Distribution")
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.show()
```

```python
#STEP 7: Word Cloud for Better Visualization
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(" ".join(filtered_words))
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title(" Word Cloud of Generated Text")
plt.show()
```

```python
#STEP 8: Stemming vs Lemmatization
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
stemmed_words = [stemmer.stem(word) for word in filtered_words]
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]

print("\n Stemming:\n", stemmed_words)
print("\n Lemmatization:\n", lemmatized_words)
```

```python
#STEP 9: Compare Original vs Stemmed vs Lemmatized
comparison_df = list(zip(filtered_words, stemmed_words, lemmatized_words))
print("\n Word Comparison (Original | Stemmed | Lemmatized):")
for original, stem, lemma in comparison_df:
    print(f"{original} | {stem} | {lemma}")
```

```python
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
from rouge_score import rouge_scorer
import numpy as np
import math
import nltk
from nltk.tokenize import word_tokenize

nltk.download('punkt')

# Sample generated text & reference text
generated_text = "The moon shines brightly over the quiet forest, whispering secrets through the leaves."
reference_text = "The moon glows softly above the silent woods, carrying whispers with the wind."

# Tokenization
reference_tokens = [word_tokenize(reference_text)]
generated_tokens = word_tokenize(generated_text)

# 1. BLEU Score with Smoothing
smoother = SmoothingFunction().method1  # Applies smoothing to avoid 0 score
bleu_score = sentence_bleu(reference_tokens, generated_tokens, smoothing_function=smoother)
print("\nBLEU Score:", round(bleu_score, 4))

# 2. ROUGE Score
rouge = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
rouge_scores = rouge.score(reference_text, generated_text)

print("\nROUGE Scores:")
print("ROUGE-1:", round(rouge_scores['rouge1'].fmeasure, 4))
print("ROUGE-2:", round(rouge_scores['rouge2'].fmeasure, 4))
print("ROUGE-L:", round(rouge_scores['rougeL'].fmeasure, 4))

# 3. Perplexity Score
def calculate_perplexity(probabilities):
    entropy = -sum(math.log(p) for p in probabilities) / len(probabilities)
    return math.exp(entropy)

probabilities = [0.2, 0.1, 0.3, 0.4]  # Replace with actual model probabilities
perplexity_score = calculate_perplexity(probabilities)
print("\nPerplexity Score:", round(perplexity_score, 4))
```

```python
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
from rouge_score import rouge_scorer
import numpy as np
import math
import nltk
from nltk.tokenize import word_tokenize

nltk.download('punkt')

# Sample generated text & multiple reference texts
generated_text = "The moon shines brightly over the quiet forest, whispering secrets through the leaves."
reference_texts = [
    "The moon glows softly above the silent woods, carrying whispers with the wind.",
    "The bright moon lights up the peaceful forest, as the trees whisper their secrets."
]

# Tokenization
reference_tokens = [word_tokenize(ref) for ref in reference_texts]
generated_tokens = word_tokenize(generated_text)

# 1. BLEU Score with Better Smoothing & Multiple References
smoother = SmoothingFunction().method4  # Stronger smoothing
bleu_score = sentence_bleu(reference_tokens, generated_tokens, smoothing_function=smoother)
print("\nBLEU Score:", round(bleu_score, 4))

# 2. ROUGE Score
rouge = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
rouge_scores = rouge.score(reference_texts[0], generated_text)  # Using first reference

print("\nROUGE Scores:")
print("ROUGE-1:", round(rouge_scores['rouge1'].fmeasure, 4))
print("ROUGE-2:", round(rouge_scores['rouge2'].fmeasure, 4))
print("ROUGE-L:", round(rouge_scores['rougeL'].fmeasure, 4))

# 3. Perplexity Score
def calculate_perplexity(probabilities):
    entropy = -sum(math.log(p) for p in probabilities) / len(probabilities)
    return math.exp(entropy)

probabilities = [0.2, 0.1, 0.3, 0.4]  # Replace with actual model probabilities
perplexity_score = calculate_perplexity(probabilities)
print("\nPerplexity Score:", round(perplexity_score, 4))
```
